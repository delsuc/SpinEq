Example of use of the SpinEq program

All examples are in examples.py

```python
import SpinEq
import examples as eg
def doc(func):
    "prints the documentation of the function"
    print(func.__doc__)
def do(func):
    "prints the documentation of the function - and run it"
    doc(func)
    print('____________________________________________\n')
    func()
```

## First the chemical equilibria

### A simple interaction 
Typically a protein (E) - Ligand (L) interaction

```python
doc(eg.EL_eq)
```

```python
eg.EL_eq(verbose=True);
```

### The effect of varying the equilibrium constant $K_{eq}$ on the previous example

```python
do(eg.EL_titr)
```

### BSA: a protein with two independent sites

```python
doc(eg.BSA)
```

```python
eg.BSA(verbose=True);
```

### Titrating the protein from the previous example

```python
do(eg.BSA_titre)
```

### A kinetic reaction

```python
do(eg.Robertson)
```

## Then check spin dynamics - (with no chemical exchange)
### 2-spins nOe

```python
do(eg.S2)
```

### saturation-recovery in a 2 spin system

```python
do(eg.T1)
```

### as above, for 5 spins

```python
do(eg.T1n)
```

### inversion-recovery in a 2 spin system

```python
do(eg.Inv2)
```

### The same on a 5 spin system,
#### first with a negative $\sigma$
corresponding to a slow tumbling molecule (protein)

```python
do(eg.Invn)
```

#### then with a positive $\sigma$
corresponding to a fast tumbling molecule

```python
def possig():
    """
    with a positive sigma
    corresponding to a fast tumbling molecule
    """
    return eg.Invn(sigma=0.3)
do(possig)
```

## Finally mix both spins and equilibrium
### Saturation Transfer Difference simulation
Using the same Protein-Ligand interaction, simulate a STD experiment by saturating the protein and looking the free ligand magnetisation

```python
do(eg.EL_STD)
```

#### evolution of EL_STD upon variation of contact time and Kd,
showing the non-linearity of the response.

```python
do(eg.EL_evol)
```
